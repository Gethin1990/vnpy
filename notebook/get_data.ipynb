```python
import akshare as aks
from datetime import datetime, timedelta

def future_history_bar(symbol, length=100):
    start_date = (datetime.now() - timedelta(days=length * 2)).strftime("%Y%m%d")
    data = aks.futures_main_sina(symbol=symbol, start_date=start_date)
    data = data.rename(columns={
    "日期": "date", "开盘价": "open", "最高价": "high",
    "最低价": "low", "收盘价": "close", "成交量": "volume",
    "持仓量": "open_interest", "动态结算价": "settle_price"
    })
    data["open"] = data["open"].astype(float)
    data["high"] = data["high"].astype(float)
    data["low"] = data["low"].astype(float)
    data["close"] = data["close"].astype(float)
    return data.iloc[-length:]

# 示例调用
symbol = "C0" # 玉米主力合约
data = future_history_bar(symbol, length=100)
print(data)
```

```python
# 创建VnPy兼容的示例数据（如果无法获取真实数据）
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os

def create_sample_data_for_vnpy():
    """创建VnPy兼容的示例数据"""
    
    # 创建日期范围
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2024, 12, 31)
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')
    
    # 过滤掉周末
    date_range = [d for d in date_range if d.weekday() < 5]
    
    # 创建示例价格数据
    np.random.seed(42)
    base_price = 4000.0
    
    data = []
    current_price = base_price
    
    for date in date_range:
        # 模拟价格变动
        change = np.random.normal(0, 20)
        current_price += change
        current_price = max(current_price, base_price * 0.8)  # 设置最低价
        
        open_price = current_price
        high_price = open_price + abs(np.random.normal(0, 15))
        low_price = open_price - abs(np.random.normal(0, 15))
        close_price = low_price + (high_price - low_price) * np.random.random()
        volume = int(np.random.normal(10000, 2000))
        
        # 确保high >= max(open, close) and low <= min(open, close)
        high_price = max(high_price, open_price, close_price)
        low_price = min(low_price, open_price, close_price)
        
        data.append({
            'datetime': date.strftime('%Y-%m-%d %H:%M:%S'),
            'open': round(open_price, 2),
            'high': round(high_price, 2),
            'low': round(low_price, 2),
            'close': round(close_price, 2),
            'volume': max(volume, 1000),  # 确保成交量大于0
            'open_interest': int(np.random.normal(50000, 5000))
        })
        
        current_price = close_price
    
    df = pd.DataFrame(data)
    return df

# 创建示例数据
print("创建VnPy兼容的示例数据...")
sample_df = create_sample_data_for_vnpy()

print("示例数据预览:")
print(sample_df.head())
print(f"\n数据形状: {sample_df.shape}")
print(f"数据类型:")
print(sample_df.dtypes)

# 保存示例数据
sample_csv_path = "data/vnpy_sample_data.csv"
os.makedirs("data", exist_ok=True)
sample_df.to_csv(sample_csv_path, index=False, encoding="utf-8-sig")

print(f"\nVnPy兼容的示例数据已保存到: {os.path.abspath(sample_csv_path)}")
print("您可以在VnPy数据管理器中导入此文件进行测试")
```

```python
# 验证CSV文件格式是否与VnPy兼容
def verify_vnpy_csv_format(csv_path):
    """验证CSV文件是否符合VnPy的导入要求"""
    try:
        df = pd.read_csv(csv_path, encoding='utf-8-sig')
        
        print(f"正在验证文件: {csv_path}")
        print(f"数据行数: {len(df)}")
        print(f"列名: {list(df.columns)}")
        
        # VnPy必需的列
        required_columns = ['datetime', 'open', 'high', 'low', 'close', 'volume']
        missing_columns = [col for col in required_columns if col not in df.columns]
        
        if missing_columns:
            print(f"❌ 缺少必需的列: {missing_columns}")
            return False
        else:
            print("✅ 所有必需的列都存在")
        
        # 检查datetime列格式
        try:
            df['datetime'] = pd.to_datetime(df['datetime'])
            print("✅ datetime列格式正确")
        except Exception as e:
            print(f"❌ datetime列格式错误: {e}")
            return False
        
        # 检查数值列
        numeric_columns = ['open', 'high', 'low', 'close', 'volume']
        for col in numeric_columns:
            if df[col].dtype not in ['int64', 'float64']:
                try:
                    pd.to_numeric(df[col])
                    print(f"✅ {col}列可以转换为数值")
                except:
                    print(f"❌ {col}列无法转换为数值")
                    return False
            else:
                print(f"✅ {col}列是数值类型")
        
        # 检查是否有空值
        null_counts = df.isnull().sum()
        if null_counts.sum() > 0:
            print("⚠️  发现空值:")
            for col, count in null_counts.items():
                if count > 0:
                    print(f"  {col}: {count}个空值")
        else:
            print("✅ 没有空值")
        
        # 显示数据预览
        print("\n数据预览:")
        print(df.head(3))
        
        print("\n✅ CSV文件格式验证通过，可以导入VnPy!")
        return True
        
    except Exception as e:
        print(f"❌ 验证失败: {e}")
        return False

# 验证生成的示例数据
print("=== 验证示例数据 ===")
verify_vnpy_csv_format("data/vnpy_sample_data.csv")

# 如果存在期权数据文件，也验证一下
option_csv_path = "data/option_cffex_hs300_daily_sina.csv"
if os.path.exists(option_csv_path):
    print("\n=== 验证期权数据 ===")
    verify_vnpy_csv_format(option_csv_path)
```
